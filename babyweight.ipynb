```python
#import dependencies
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_file = Path("Resources\\us_births_2016_2021.csv")
```

```python
us_births_df = pd.read_csv("us_births_2016_2021.csv")
us_births_df
```

```python
us_births_df['Average Age of Mother (years)'] = us_births_df['Average Age of Mother (years)'].astype(float).astype(int)
us_births_df
```

```python
mothers_age = us_births_df.groupby(["Average Age of Mother (years)"])['Average Birth Weight (g)'].count()
mothers_age
```

```python
bby_plot = mothers_age.plot(kind="bar", color='b', figsize=(10,8), fontsize = 14)
plt.title = ("Average Age of Mother vs Birthweight (g)")
plt.xlabel("Age",fontsize = 15)
plt.ylabel("Weight (grams)",fontsize = 15)

ticks = range(0, 23, 36)
labels = ticks
plt.xticks(ticks, labels)
                           
plt.show()
plt.tight_layout()

bby_plot
```
