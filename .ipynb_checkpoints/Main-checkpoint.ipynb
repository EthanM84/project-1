```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt

# Specify the path to your CSV file
csv_file_path = 'resources/births_data.csv'

# Read the CSV file into a DataFrame
data_df = pd.read_csv(csv_file_path)

# Print the DataFrame
data_df.head()
```

```python
# Count the total number of entries in the DataFrame
column_entries_count = data_df.count()

column_entries_count
```

```python
# Calculate the standard deviation for the "Number of Births" column
births_std = data_df['Number of Births'].std()

# Calculate the standard deviation for the "Average Birth Weight (g)" column
weight_std = data_df['Average Birth Weight (g)'].std()

print("Standard Deviation of Number of Births:", births_std)
print("Standard Deviation of Average Birth Weight (g):", weight_std)

# Calculate the mode for the "Number of Births" column
births_mode = data_df['Number of Births'].mode()

# Calculate the mean for the "Number of Births" column
births_mean = data_df['Number of Births'].mean()

print("Mode of Number of Births:", births_mode)
print("Mean of Number of Births:", births_mean)
```

```python
# Calculate the average age of the mother per state per year
average_age = data_df.groupby(['State', 'Year'])['Average Age of Mother (years)'].mean()
# Convert the result to a DataFrame
average_age_df = average_age.reset_index()
# Printing it
average_age_df
```

```python
# Calculate the average age of the mother per state
average_age_per_state = data_df.groupby('State')['Average Age of Mother (years)'].mean()

average_age_per_state
```

```python
# Visualize the average age per state using a bar plot
plt.figure(figsize=(10, 6))
average_age_per_state.plot(kind='bar')
plt.xlabel('State')
plt.ylabel('Average Age of Mother (years)')
plt.title('Average Age of Mother per State')
plt.tight_layout()
plt.show()
```

```python
# Calculate the average weight per gender
average_weight_per_gender = data_df.groupby('Gender')['Average Birth Weight (g)'].mean()
average_weight_per_gender
```

```python
# Visualize the average weight per gender using a bar plot
plt.figure(figsize=(6, 4))
average_weight_per_gender.plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Average Birth Weight (g)')
plt.title('Average Birth Weight per Gender')
plt.tight_layout()
plt.show()
```
